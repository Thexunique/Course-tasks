```python
# Imports
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import arabic_reshaper
from bidi.algorithm import get_display
```

```python
# Load Arabic font
font_path = 'amiri/Amiri-0.113/Amiri-Bold.ttf'
arabic_font = fm.FontProperties(fname=font_path)
```

```python
# Load Arabic font
font_path = 'amiri/Amiri-0.113/Amiri-Regular.ttf'
arabic_font = fm.FontProperties(fname=font_path)
```

```python
# Arabic day names (reshaped and RTL corrected)
raw_days = ['الاثنين', 'الثلاثاء', 'الأربعاء', 'الخميس', 'الجمعة', 'السبت', 'الأحد']
days_arabic = [get_display(arabic_reshaper.reshape(day)) for day in raw_days]
temperatures = [22, 24, 19, 23, 25, 27, 21]
```

```python
# Plotting
plt.figure(figsize=(9, 5))
plt.plot(days_arabic, temperatures, marker='o', linestyle='-', color='teal')
# Axis labels and title (also reshaped)
plt.title(get_display(arabic_reshaper.reshape('تغير درجة الحرارة خلال الأسبوع')), fontproperties=arabic_font)
plt.xlabel(get_display(arabic_reshaper.reshape('أيام الأسبوع')), fontproperties=arabic_font)
plt.ylabel(get_display(arabic_reshaper.reshape('درجة الحرارة (°م)')), fontproperties=arabic_font)
plt.grid(True)
plt.tight_layout()
plt.show()
```
