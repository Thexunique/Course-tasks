Define individual item prices

```python
price_bluetooth_speaker =110.5
price_wireless_mouse = 125.6
price_desk_lamp = 50
```

# Define the budget available for shopping


```python
while True:
    try:
        budget = float(input("Enter your total budget for shopping: EG"))
        break
    except ValueError:
        print("Please enter a valid number (e.g., 100.00).")
```

Calculate the total cost of the selected items

```python
total_purchase_amount = price_bluetooth_speaker + price_wireless_mouse + price_desk_lamp
```

Compare the total cost against the available budget

```python
if total_purchase_amount <= budget:
    leftover_funds = budget - total_purchase_amount
    print(f"Purchase successful! You have EG{leftover_funds:.2f} left in your budget.")
else:
    extra_money_needed = total_purchase_amount - budget
    print(f"Not enough funds! You need an additional EG{extra_money_needed:.2f} to complete the purchase.")
```

This is Extra

```python
print(f"\n=== Purchase Summary ===")
print(f"Total cost of items: EG{total_purchase_amount:.2f}")
print(f"Budget available:    Eg{budget:.2f}")
```
